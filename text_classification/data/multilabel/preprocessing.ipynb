```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('stores_data.csv',lineterminator='\n')
df.info()
```

```python
df.drop(['Unnamed: 0', 'store_title','store_collections'], axis=1, inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
df['store_labels'] = df['store_labels'].astype(str)
clothing_stores = df[df['store_labels'].str.contains('car')]
clothing_stores.head()

label_counts = Counter()
for labels in clothing_stores['store_labels']:
    labels = labels[1:-1]  # remove leading and trailing brackets
    label_counts.update(labels.split(', '))
print(len(label_counts))
print(clothing_stores.info())
```
